```python
import spacy
nlp=spacy.load("en_core_web_lg")
```

```python
doc = nlp("The sun dipped below the horizon, casting hues of orange and pink across the sky. The gentle breeze rustled the leaves of the trees, creating a soothing melody that seemed to dance with the fading light. As twilight descended, the world transformed into a realm of magic and mystery. Shadows lengthened, blending with the soft glow of lanterns that dotted the pathways. In this fleeting moment between day and night, anything felt possible. It was a time to reflect on the beauty of nature and to embrace the tranquility of the evening.")
```

```python
for word in doc:
  print(word)
```

```python
sents=[]
for sent in doc.sents:
  sents.append(sent)
```

```python
sents
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
list1=[]
for word in doc:
  if word.is_stop==False:
    list1.append(word.text)

docn=" ".join(list1)
```

```python
docn
```

```python
for word in doc:
  print(word)
```

```python
docn=nlp(docn)
```

```python
sents=[]
for sent in docn.sents:
  sents.append(sent)
```

```python
sents
```

```python
sentence_vectors = []
for sentence in sents:
    vector = sentence.vector
    sentence_vectors.append(vector)
```

```python
sentence_vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(sentence_vectors)

print("Similarity Matrix:")
print(similarities)
```

```python
threshold = 0.8
for i in range(len(sents)):
    for j in range(i + 1, len(sents)):
        if similarities[i][j] > threshold:
            print(f"Sentence {i+1}:", sents[i])
            print(f"Sentence {j+1}:", sents[j])
            print("Cosine Similarity:", similarities[i][j])
            print()
```
